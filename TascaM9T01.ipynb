```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
nltk.download('stopwords')
```

### Exercici 1
Agafa un text en anglès que vulguis, i calcula'n la freqüència de les paraules.

```python
# Load your text file
with open('Housman.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
tokens = word_tokenize(text)

# Remove non-alphabetic words and convert to lowercase
filtered_tokens = [token.lower() for token in tokens if token.isalpha()]

# Create a FreqDist object for filtered words
filtered_words = FreqDist(filtered_tokens)

# Print the most common words after filtering
print(filtered_words.most_common(10))

# Plot the frequency distribution
filtered_words.plot()
```

### Exercici 2
Treu les stopwords i realitza stemming al teu conjunt de dades.

```python
# Get the English stop words
stop_words = set(stopwords.words('english'))

# Create a new list of filtered words excluding stop words
filtered_words = [word for word in filtered_tokens if word not in stop_words]

# Create a FreqDist object for the filtered words without stopwords
freq_dist = FreqDist(filtered_words)

# Print the frequency distribution without stopwords
print(freq_dist)
filtered_words
```

### Exercici 3
Realitza sentiment analysis al teu conjunt de dades.

```python
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()


# Create a function to get sentiment
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return sentiment

# Apply the get_sentiment function to your text
sentiment = get_sentiment(text)

# Print the sentiment result
print("Sentiment:", sentiment)
```

El resultat indica que el sentiment del text és positiu.
